```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 데이터 불러오기
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train, y_test = to_categorical(y_train), to_categorical(y_test)

# 모델 구성
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 학습 (1 에폭만 테스트용)
model.fit(x_train, y_train, epochs=1, batch_size=128, validation_split=0.1)
```

```python
import tensorflow as tf

# 예제 데이터: (1, 2, 3) == (batch=1, 2개 엔티티, 임베딩 차원=3)
v_embed = tf.constant([[[1., 2., 3.],
                        [4., 5., 6.]]])  # (1, 2, 3)

r_embed = tf.constant([[[10., 20., 30.],
                        [40., 50., 60.]]])  # (1, 2, 3)
# print(v_embed.shape)
# print(r_embed.shape)

# 차원 확장
v_expand = tf.expand_dims(v_embed, axis=2)  # (1, 2, 1, 3)
r_expand = tf.expand_dims(r_embed, axis=1)  # (1, 1, 2, 3)

print(v_expand)
# print(v_expand.shape)
# print(r_expand.shape)

# 타일 복제
v_tiled = tf.tile(v_expand, [1, 1, 2, 1])  # (1, 2, 2, 3)
r_tiled = tf.tile(r_expand, [1, 2, 1, 1])  # (1, 2, 2, 3)
print(v_tiled)

# 마지막 차원 기준으로 붙이기
pair_embed = tf.concat([v_tiled, r_tiled], axis=-1)  # (1, 2, 2, 6)

# 결과 확인
print("v_tiled:\n", v_tiled.numpy())
print("r_tiled:\n", r_tiled.numpy())
print("pair_embed:\n", pair_embed.numpy())
```
